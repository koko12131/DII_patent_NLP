```python
import networkx as nx
import difflib
```

```python
G = nx.MultiGraph()
for x in data['AE']:
  if len(eval(x))>1:
    for y in eval(x):
      for z in eval(x):
        if y!=z and difflib.SequenceMatcher(lambda p:len(p)<5,y,z).quick_ratio()<0.8:
          print([y,z])
          G.add_edge(y,z)
nx.write_gexf(G,'./patenttee_corporation.gexf')
```
